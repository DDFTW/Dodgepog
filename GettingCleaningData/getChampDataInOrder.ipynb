```python
champArray = []

for match in matches:
    
    #get the 10 champions, in order by participants
    #store these 10 in an array
    for participant in match['participants']:
        identity = participant['championId']
        
        #convert to champ name?
        name = get_champions_name(identity)
        #add name to array
        champArray.append( name )

    champArray.append( int(match['team1Win']) )
    

champNames = np.asarray(champArray)
champNames = champNames.reshape( (-1,11))
print(champNames)
```

```python
#convert array to dataframe
cat = []
for i in range(1,11):
    cat.append('player{}'.format(i))
cat.append('team1Win')
dataframe = pd.DataFrame(data=champNames, columns = cat)
dataframe.head()
```
